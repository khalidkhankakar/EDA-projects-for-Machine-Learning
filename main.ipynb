```python
from sympy.plotting import plot3d_parametric_surface

u, v = symbols('u v')
r = 2

x = r * cos(u) * sin(v)
y = r * sin(u) * sin(v)
z = r * cos(v)

plot3d_parametric_surface(x, y, z, (u, 0, 2*pi), (v, 0, pi),
                          title="Sphere of radius 2")
```

```python
from sympy import symbols
from sympy.plotting import plot3d_parametric_surface

u, v = symbols('u v')

a, b, c = 2, 3, 4  # box dimensions

# 6 faces of the rectangle
faces = [
    (0, u, v, (u, 0, b), (v, 0, c)),   # x=0
    (a, u, v, (u, 0, b), (v, 0, c)),   # x=a
    (u, 0, v, (u, 0, a), (v, 0, c)),   # y=0
    (u, b, v, (u, 0, a), (v, 0, c)),   # y=b
    (u, v, 0, (u, 0, a), (v, 0, b)),   # z=0
    (u, v, c, (u, 0, a), (v, 0, b)),   # z=c
]

# Plot all faces
plots = []
for face in faces:
    plots.append(plot3d_parametric_surface(face[0], face[1], face[2],
                                           face[3], face[4],
                                           show=False))

# Combine into one 3D plot
p = plots[0]
for q in plots[1:]:
    p.extend(q)
p.show()
```

```python
from sympy import symbols
from sympy.plotting import plot_parametric

t = symbols('t')

# Rectangle edges (parametric)
edges = [
    (t, 0, (t, 0, 5)),   # bottom edge (0,0) → (4,0)
    (5, t, (t, 0, 2)),   # right edge (4,0) → (4,2)
    (t, 2, (t, 0, 5)),   # top edge (4,2) → (0,2)
    (0, t, (t, 0, 2))    # left edge (0,2) → (0,0)
]

# Plot all edges
plots = []
for x, y, param in edges:
    plots.append(plot_parametric(x, y, param, show=False))

# Combine into one plot
p = plots[0]
for q in plots[1:]:
    p.extend(q)
p.show()
```
